```python
import streamlit as st
import pandas as pd
import joblib

# Page Configuration

st.set_page_config(
    page_title="Fraud Prediction Model",
    page_icon="🤖",
    layout="centered"
)


# Load Model and Scaler

model = joblib.load("fraud_detection_model.pkl")
scaler = joblib.load("scaler.pkl")


# Title

st.title("💳 Fraud Prediction System")
st.markdown("Predict whether a customer is likely to **Fraud on credit payment**.")

st.divider()


# Sidebar

st.sidebar.header("About")

st.sidebar.info(
    """
    **Machine Learning Project**

    Predicts whether a customer is likely to Fraud based on:

    - Student Status
    - Credit Balance
    - Annual Income

    Built using:
    - Streamlit
    - Scikit-Learn
    - Python
    """
)


# User Inputs

student = st.selectbox(
    "Student Status",
    ["No", "Yes"]
)

balance = st.number_input(
    "Account Balance",
    min_value=0.0,
    value=1000.0,
    step=100.0
)

income = st.number_input(
    "Annual Income",
    min_value=0.0,
    value=30000.0,
    step=1000.0
)


# Prediction

if st.button("Predict Fraud"):

    student_encoded = 1 if student == "Yes" else 0

    input_data = pd.DataFrame({
        "student": [student_encoded],
        "balance": [balance],
        "income": [income]
    })

    # Scale numerical features
    input_data[["balance", "income"]] = scaler.transform(
        input_data[["balance", "income"]]
    )

    prediction = model.predict(input_data)[0]

    probability = model.predict_proba(input_data)[0]

    st.divider()

    st.subheader("Prediction Result")

    if prediction == 1:
        st.error("⚠️ Customer is likely to Fraud.")
    else:
        st.success("✅ Customer is NOT likely to Fraud.")

    st.write("### Prediction Confidence")

    st.progress(float(max(probability)))

    st.write(f"Probability of Not a Fraud : **{probability[0]*100:.2f}%**")
    st.write(f"Probability of Fraud : **{probability[1]*100:.2f}%**")

    st.divider()

    st.subheader("Customer Information")

    st.write(f"**Student:** {student}")
    st.write(f"**Balance:** ₹{balance:,.2f}")
    st.write(f"**Income:** ₹{income:,.2f}")

# --------------------------------------------------
# Footer
# --------------------------------------------------
st.divider()

st.markdown(
    """
    ---
    **Developed by:** Anshika Chauhan  
    **Project:** Fraud Prediction using Machine Learning
    """
)
```
